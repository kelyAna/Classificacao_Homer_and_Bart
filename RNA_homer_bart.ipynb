**Rede Neural densa, para classificar imagens dos personagens Homer e Bart**


Importações iniciais

```
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

Leitura da base de dados


```
dataFrame = pd.read_csv('personagens.csv')
```

```
dataFrame
```

```
X = dataFrame.iloc[:,0:6].values
```

```
X
```

```
y = dataFrame.iloc[:,6].values
```

```
y
```

Transformar as labels categóricas em números

```
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

Divisão do dataFrame em treino e teste


*   train_test_split



```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```
len(X_train), len(X_test), len(y_train), len(y_test)
```

Estrutura da Rede Neural Densa

```
classificador = Sequential()
```

```
classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform',
                          input_dim = 6))
```

```
classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'))
```

```
classificador.add(Dense(units = 1, activation = 'sigmoid'))
```

Configuração da Rede Neural

```
otimizador = keras.optimizers.Adam(learning_rate = 0.0001, decay = 0.0001, clipvalue = 0.5)
```

```
classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
```

Treinamento da Rede Neural

```
classificador.fit(X_train, y_train, batch_size = 5, epochs =1000)
```

```
previsoes = classificador.predict(X_test)
```

```
previsoes = (previsoes > 0.5)
```

```
previsoes
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
```

```
precisao = accuracy_score(y_test, previsoes)
```

```
precisao
```

```
matriz = confusion_matrix(y_test, previsoes)
matriz
```
